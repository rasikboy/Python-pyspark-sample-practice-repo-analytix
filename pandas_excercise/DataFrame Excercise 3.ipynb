# isnull() and notnull() in python pandas

```python
mask = df["Team"].isnull() 
#above line will return the true where it will get null else will be False
df[mask]
```

```python
mask = df["Team"].notnull()  # this method is just vice versa of isnull() .
df[mask]
```

# drop_duplicates Method

```python
df=pd.read_csv("D:\\Pandas_Data\\data\\employees.csv", parse_dates=["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("bool")
df.sort_values("First Name", inplace = True)
```

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(subset=["First Name"]) # inplace=True will make change permanent in the datsframe 
# keep = "First/Last" will keep either first or last value from the group of duplicate on
```
