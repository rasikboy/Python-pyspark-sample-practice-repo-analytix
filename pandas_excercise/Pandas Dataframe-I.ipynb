```python
nba.sort_values("birth_city", ascending=False, na_position="last")
# the na_position helps us to keep NaN valus in the last , 
#it is by default last, but for the very first time dataframe
# when we are doing any operation like this , we have to call it for keeping again in last these NaN values for current output
```

```python
# we can sort data based on multiple columns as in using sort_values() method, 
# we also can pass multiple ascending orders as well if we have multiple columns for the sorting
nba.sort_values(["Player","birth_state" ], na_position="last", ascending=[False,True], inplace=True)
nba
```
